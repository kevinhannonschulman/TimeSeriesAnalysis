The travel industry was perhaps the most directly impacted sector of the economy during the pandemic as international flights abruptly ground to a halt in March 2020 and many domestic carriers significantly reduced their flights. I was curious how the travel industry could rebound this year (2021) so I created a time series forecast using the FBProphet package to predict the number of passengers being screened in U.S. airports this year and specifically sought to find how the total number of passengers screened this year will compare to pre-pandemic levels in 2019.

```python
#import necessary packages
import os
import psycopg2
import pandas as pd
import numpy as np
from dotenv import load_dotenv
from prophet import Prophet
from matplotlib import pyplot as plt
```

```python
load_dotenv("database.env") #load .env file from directory
```

I used the psycopg2 package to retrieve the data from a postgres database within a Docker container and performed the analysis using a Python virtual environment within a connected Docker container. 

```python
#connect to database using psycopg2
def get_con():
    con = psycopg2.connect(
        host='0.0.0.0',
        database=os.environ.get("POSTGRES_DB"), #grab database login variables from .env
        user=os.environ.get("POSTGRES_USER"),
        password=os.environ.get("POSTGRES_PASSWORD")
    )
    cursor = con.cursor() #activate cursor

    return con, cursor
```

```python
con, cursor = get_con()
df = pd.read_sql("select date, total from public.my_table", con) #execute SQL query and read to dataframe
```

```python
df.columns = ['ds', 'y'] #rename columns for prophet
df['ds']=pd.to_datetime(df['ds']) #convert date column to datetime
df['y'] = df['y'].astype(int) #convert number of passengers (y) to integers
```

```python
df.head()
```

```python
df.plot(x='ds', y='y', xlabel='Date', ylabel='Passengers Screened (millions)',
        title='Passengers Screened at U.S. Airports Per Day (Jan 2019 â€“ Apr 2021)', figsize = (10,6))
#visualize data to ensure accuracy before running model
```

```python
# declare model and fit to dataframe
m = Prophet()
m.fit(df)
#create dataframe for one year into future
future = m.make_future_dataframe(periods=365, include_history=True)
forecast = m.predict(future) #run model on future dates
fig = m.plot(forecast, xlabel = 'Date', ylabel = 'Passengers (millions)')
ax = fig.gca()
ax.set_title("Passengers Screened at U.S. Airports Per Day")
for x in m.changepoints:
    plt.axvline(x, c='red', ls='--', lw=1) #plot model changepoints
```

First, I declared the fbprophet model (similar to sklearn syntax) and then fit the model to my original data. Next, I created a separate dataframe with 365 future days and used the model to predict the number of screened passengers for one year into the future. Overall, the model predicts a relatively steady rebound in the number of passengers screened over the next year. 

However, the model appears to assign significant weight to the seasonal dip in demand around March 2019 and the precipitous decline in March 2020. It is uncertain whether this pattern will necessarily repeat itself to such a significant degree next year although it predicts a similar decline around March-April 2022. Additionally, the model has made conservative predictions in the short-term exemplified below by the predictions for the week of April 26 â€“ May 3. In reality, there were 1,335,535 passengers screened on May 1 compared to the 513,774 predicted by the model and significantly greater than even the upper bound (959,733) of the prediction.

```python
springforecast = (forecast['ds'] >= '2021-04-26') & (forecast['ds'] <= '2021-05-03')
forecast.loc[springforecast] #predicted number of passengers during week of 4/26â€“5/3
```

```python
plot2 = m.plot_components(forecast)
```

FBProphet automatically deconstructs the data into various seasonality components which show that there is clear weekly and yearly seasonality in the data. Specifically, the fewest passengers are screened on Tuesdays which makes sense because few people except business travelers are flying during the week. Conversely, the greatest number of passengers are screened on Thursdays and Fridays as passengers travel over the weekend and again on Sunday as they return home before the beginning of the week. Annually, the data is more complicated because I believe the model is still assigning too much significance to the March-May decline seen in 2019 and 2020. Travel does rebound during the summer months as more people take vacations before decreasing during early fall as school resumes and rising during the holiday season, especially around Christmas.

```python
deltas = m.params['delta'].mean(0)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111)
ax.bar(range(len(deltas)), deltas, color='red')
ax.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2)
ax.set_ylabel('Rate change')
ax.set_xlabel('Potential changepoint')
ax.set_title("Rate Change Impact of Changepoints")
fig.tight_layout()
```

The model automatically determines 25 potential changepoints in the first 80% of the time series and I was curious how each of these changepoints affected the model. The first five changepoints encompass spring 2019 and the minor decline in screened passengers which then rebounds in the next few months and leads to almost no rate change. Interestingly, the next significant changepoints occur mid-November 2019 to early January 2020 and, visually, there is a clear dip in the forecast trend line around this time. Around this time, news of the impact of the virus in China was beginning to spread and so it is possible that some people were more hesitant to travel beginning during this period. A nationwide travel ban from China began on January 31, 2020 but it is unlikely this solely accounts for the change because international travelers are not included in this data unless they connected at a U.S. airport on their way to another domestic destination. It is possible that Last ones August to September 2020. Lastly, the final significant changepoints occur in August/September 2020 as travel began to slowly resume.

```python
#evaluate model accuracy
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

metric_df = forecast.set_index('ds')[['yhat']].join(df.set_index('ds').y).reset_index() #join original and forecast data
metric_df.dropna(inplace=True) #drop null values
print(r2_score(metric_df.y, metric_df.yhat), mean_squared_error(metric_df.y, metric_df.yhat, squared=False), 
      mean_absolute_error(metric_df.y, metric_df.yhat)) #calculate R-Squared, MSE, MAE
```

# Accuracy metrics

We can use the metrics within the sklearn package to analyze the accuracy of the default model without any optimization. Below are some thoughts about each metric:

- The R-Squared value of 93% implies that 93% of the variance is explained by the model which seems good but could also indicate overfitting. FBProphet uses a linear model for its default forecast but a nonlinear model may be more accurate for this data.  

- The Root Mean Squared Error (RMSE) value indicates that the model misses actual passenger numbers by 221,832. However, this metric is highly sensitive to outliers because the error is squared which likely had a significant impact with such volatile data.

- Similarly, the Mean Absolute Error (MAE) value of 161,344 seems large (nearly 400 full Boeing 747s) but is reasonable when the total number of passengers screened per day is measured in millions. MAE does not square each error and is less sensitive to outliers than RMSE as a result.

```python
#calculate difference in total passengers for 2019 vs. 2021
mask = (df['ds'] >= '2019-01-01') & (df['ds'] <= '2019-12-31')
df.loc[mask].sum() #total annual passengers for 2019 calendar year

mask2 = (forecast['ds'] >= '2021-01-01') & (forecast['ds'] <= '2021-12-31')
forecast['yhat'].loc[mask2].sum() #total annual passengers for 2021 calendar year

print(int(df.loc[mask].sum()) - forecast['yhat'].loc[mask2].sum())
```

# Conclusions

Finally, we can use the results of the time series forecast to answer how many fewer people are predicted to be screened at U.S. airports this year compared to pre-pandemic levels in 2019. According to the model, roughly 420,000,000 fewer people will fly through U.S. airports during the 2021 calendar year than the 2019 calendar year though, as noted earlier, the model seems to take a conservative approach so it is possible that this difference will actually be smaller.

Possible Improvements:

I ran the model with the default settings so it is likely possible to achieve more accurate results by tuning certain model parameters such as seasonality and changepoints. The FBProphet package includes cross-validation features that can be used to find the best values for each parameter. Additionally, I only used one complete year (2019) of pre-pandemic data so the forecast would likely be more accurate if the model was trained on multiple years of past data to reflect the broader industry growth over the past decade.

Further Questions:

As noted before, this data only includes passengers screened at U.S. airports so it primarily focuses on domestic travel. It would be interesting to analyze international data as well to see how travel patterns differ between  countries as international flights begin to resume.
